# Для 1 переменной

```python
import sympy as sp
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
x = sp.symbols('x')
```

```python
f_x = 3*x**2
```

```python
df_x = sp.diff(f_x, x)
```

```python
x0 = 5
lam = 0.1
```

```python
res = []
while True:
    res.append([x0, f_x.subs(x, x0) ])
    x1 = x0 - lam * df_x.subs(x, x0)
    if math.sqrt((abs(x0-x1))**2) < 0.00000001:
        break
    x0 = x1
```

```python
x_values = np.linspace(-10, 10, 400)
f_values = [f_x.subs(x, val).evalf() for val in x_values]

plt.figure(figsize=(10, 6))
plt.plot(x_values, f_values, label='f(x) = 3*x^2')
plt.scatter([x[0] for x in res], [x[1] for x in res], color='red', label='Точки градиентного спуска')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('График функции и точек градиентного спуска')
plt.legend()
plt.grid(True)
plt.show()
```

# Для 2 переменных

```python
x = sp.symbols('x')
y = sp.symbols('y')
```

```python
f = 3*x**2 + 5*y**2
```

```python
df_x = sp.diff(f, x)
df_y = sp.diff(f, y)
```

```python
x0, y0 = 5, 10
lam = 0.1
```

```python
res = []

while True:
    res.append([x0, y0, f.subs({x: x0, y: y0})])
    x1 = x0 - lam * df_x.subs({x: x0, y: y0})
    y1 = y0 - lam * df_y.subs({x: x0, y: y0})

    if  math.sqrt((abs(x0-x1))**2 + (abs(y0-y1))**2) < 0.00000001:
        break

    x0, y0 = x1, y1
```

```python
res
```

```python
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

X = np.linspace(-10, 10, 100)
Y = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(X, Y)
Z = 3*X**2 + 5*Y**2

ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8)  # Построение поверхности функции

# Отмечаем точки градиентного спуска на графике
for point in res:
    ax.scatter(point[0], point[1], point[2], color='red', s=50)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')
ax.set_title('3D график функции и точек градиентного спуска')

plt.show()
```
